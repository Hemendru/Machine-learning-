```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv(r"C:\Users\HP\Machines\titanic.csv")

print(df.head(3))
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python


plt.figure(figsize=(8,6))
sns.boxplot(x='Embarked',y='Fare',data=df)
plt.title("Fare Distribution by Embarked Port")
plt.show()

```

```python
fig,axes = plt.subplots(1,2,figsize=(12,5))
sns.histplot(df['Age'],bins=20,kde=True,color='orange',ax=axes[0])

sns.histplot(df['Fare'],bins=20,kde=True,color='red',ax=axes[1])
axes[0].set_title('Histogram of Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('count')

axes[1].set_title('histogram of Fare')
axes[1].set_xlabel('Fare')
axes[1].set_ylabel('Count')
plt.tight_layout()
plt.show()
```
